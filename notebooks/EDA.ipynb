## Handling Missing Values

The dataset contains meaningful missing values. Instead of dropping these rows, we replace them with appropriate labels based on business understanding.

```python
df["Offer"] = df["Offer"].fillna("No Offer")
```

```python
df["Internet Type"] = df["Internet Type"].fillna("No Internet")
```

```python
df["Churn Category"] = df["Churn Category"].fillna("Not Churned")
```

```python
df["Churn Reason"] = df["Churn Reason"].fillna("Not Churned")
```

```python
df.isnull().sum()
```

# Target Variable Analysis

The target variable is **Churn Label**, which indicates whether a customer has churned (Yes) or stayed (No). Understanding its distribution helps identify class imbalance before model training.

# Target Variable Visualization

Visualizing the target variable helps understand the class distribution before model training.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,4))
sns.countplot(x="Churn Label", data=df)
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number Of Customers")

plt.show()
```
